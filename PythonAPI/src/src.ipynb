```python
def create_timer_slices(data):
    timer_slices = []
    current_timer_val = data[0, 0]  # Initialize current timer value to the first frame's timer value
    current_timer_slice = [data[0]]  # Initialize current timer slice with the first frame
    for frame in data[1:]:
        if frame[0] >= current_timer_val:
            current_timer_slice.append(frame)  # Add frame to current timer slice
            current_timer_val = frame[0]
        else:
            timer_slices.append(np.array(current_timer_slice))  # Append current timer slice to timer slices
            current_timer_val = frame[0]
            current_timer_slice = [frame]  # Start new timer slice with the current frame
    timer_slices.append(np.array(current_timer_slice))  # Append final timer slice to timer slices
    return timer_slices

def create_window_slices(timer_slices, window_size):
    window_slices = []
    for timer_slice in timer_slices:
        num_frames = len(timer_slice)
        if num_frames >= window_size:
            for i in range(num_frames - window_size + 1):
                window_slice = timer_slice[i:i+window_size]
                # make dataframes
                window_slice = pd.DataFrame(window_slice, columns=['timer', 'opponent_character', 'player_character', 'opponent_jumping', 'opponent_crouching', 'opponent_in_move', 'opponent_Up', 'opponent_Down', 'opponent_Right', 'opponent_Left', 'opponent_Y', 'opponent_B', 'opponent_X', 'opponent_A', 'opponent_L', 'opponent_R', 'player_jumping', 'player_crouching', 'player_in_move', 'player_Up', 'player_Down', 'player_Right', 'player_Left', 'player_Y', 'player_B', 'player_X', 'player_A', 'player_L', 'player_R', 'opponent_health', 'player_health', 'x_diff', 'y_diff'])
                window_slices.append(window_slice)
    return window_slices

timer_slices = create_timer_slices(df.values)

window_slices = create_window_slices(timer_slices, 3)

window_slices[0]
```
